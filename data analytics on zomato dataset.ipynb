```python
plt.figure(figsize=(10, 6))
sns.barplot(data=ratings, x='Aggregate rating', y='Rating count', hue='Rating color')
plt.title('Bar Plot of Aggregate Rating vs. Rating Count by Rating Color')
plt.xlabel('Aggregate Rating')
plt.ylabel('Rating Count')
```

```python
plt.figure(figsize=(8, 8))
plt.pie(a[:3], labels=b[:3], autopct='%1.1f%%',colors=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Pie Chart Example')
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Has Table booking', data=df)
plt.title('Count of Restaurants with Table Booking')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=df, hue='Has Online delivery', palette='coolwarm')
plt.title('Scatterplot of Aggregate Rating by Votes')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Rating color', data=df, order=df['Rating color'].value_counts().index, palette='pastel')
plt.title('Count of Restaurants by Rating Color')
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='Price range', y='Aggregate rating', data=df, palette='viridis')
plt.title('Violin Plot of Aggregate Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(df[['Price range', 'Aggregate rating', 'Votes', 'Has Online delivery']], hue='Has Online delivery', palette='Set1')
plt.suptitle('Pair Plot of Numerical Columns with Online Delivery', y=1.02)
```

```python
plt.figure(figsize=(10, 6))
sns.stripplot(x='Has Table booking', y='Aggregate rating', data=df, palette='Set2')
plt.title('strip Plot of Aggregate Rating by Table Booking')
plt.xlabel('Has Table Booking')
plt.ylabel('Aggregate Rating')
```
